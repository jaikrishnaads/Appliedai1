```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()

image_path = list(uploaded.keys())[0]

img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
blur = cv2.GaussianBlur(gray, (5,5), 0)

plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
```

```python
edges = cv2.Canny(blur, 50, 150)

plt.imshow(edges, cmap='gray')
plt.title("Edges Detected")
plt.axis('off')
```

```python
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print("Total objects detected:", len(contours))
```

```python
output = img_rgb.copy()
cv2.drawContours(output, contours, -1, (0,255,0), 2)

plt.imshow(output)
plt.title("Detected Objects")
plt.axis('off')
```

```python
for i, cnt in enumerate(contours):
    x, y, w, h = cv2.boundingRect(cnt)

    cv2.rectangle(output, (x,y), (x+w, y+h), (255,0,0), 2)

    cv2.putText(output, f"{w}x{h}", (x, y-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)

plt.imshow(output)
plt.title("Measured Objects")
plt.axis('off')
```

```python
print("Final Object Count:", len(contours))

for i, cnt in enumerate(contours):
    x, y, w, h = cv2.boundingRect(cnt)
    print(f"Object {i+1}: Width={w}, Height={h}")
```

```python
# Object Measurement and Counting using Laptop Camera (OpenCV)

## Overview

This project implements a real-time object measurement and counting system using a laptop camera. It detects objects, draws bounding boxes, and estimates their width and height in centimeters using a reference object.

This is a rule-based computer vision pipeline built using OpenCV. It does not use machine learning.

---

## Objective

* Detect objects from a live camera feed
* Count the number of objects
* Measure object dimensions in real-world units
* Demonstrate a simple, explainable vision pipeline

---

## Key Assumptions

* A reference object with known width is present in the frame
* Camera position remains stable
* Objects are placed flat on a surface
* Lighting conditions are consistent

---

## Tech Stack

* Python
* OpenCV
* NumPy
* Matplotlib
* Google Colab (for webcam capture)

---

## Pipeline Steps

### 1. Image Capture

* Capture frame from laptop camera using browser interface
* Convert base64 image to OpenCV format

### 2. Grayscale Conversion

* Reduce image from 3 channels to 1
* Simplifies processing
* Removes unnecessary color information

### 3. Noise Reduction (Gaussian Blur)

* Smooth image
* Reduce false edge detection
* Improve contour quality

### 4. Edge Detection (Canny)

* Detect object boundaries
* Convert image to binary edge map
* White = edges, Black = background

### 5. Contour Detection

* Identify closed shapes
* Each contour represents a potential object

### 6. Reference Object Selection

* Largest contour assumed as reference
* Known real-world width used for scaling

### 7. Pixel to Real-World Conversion

* Compute pixel-per-cm ratio
* Convert bounding box dimensions into centimeters

### 8. Bounding Box and Measurement

* Draw rectangle around each object
* Display width and height

### 9. Object Counting

* Count number of valid contours

---

## Output

* Annotated image with bounding boxes
* Object count
* Width and height of each object in cm

---

## Example Use Case

* Measure size of a mobile phone
* Count number of items on a table
* Simple inspection system

---

## Limitations

* Cannot measure depth or thickness
* Accuracy depends on camera angle
* Overlapping objects reduce accuracy
* Sensitive to lighting and shadows
* Reference object detection is hardcoded

---

## Why This Approach

* Simple and fast
* due to my resources are limited
* No training data required
* Works in constrained environment

---

## Not Used (and Why)

### Sliding Window

* Slow and unnecessary
* Designed for object detection, not measurement

### Vertical/Horizontal Line Detection

* Detects internal edges, not object boundaries
* Adds noise and complexity

---

## Future Improvements

* Use ArUco markers for automatic scaling
* Add camera calibration for better accuracy
* Use depth sensors for 3D measurement
* Improve contour filtering logic

---

## Conclusion

This project demonstrates a basic computer vision pipeline for object detection and measurement using classical techniques. It focuses on clarity, speed, and explainability rather than accuracy or scalability.

---
```
