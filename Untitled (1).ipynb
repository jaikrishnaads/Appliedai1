```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r"C:\Users\jaikrishna\Downloads\Week 1-20251229T080409Z-1-001\Week 1\heart.csv")
df.head()

df.shape
df.dtypes
df.isnull().sum()

# Features and target
X = df.drop(columns=['output'])
y = df['output']

print(X)
print(X.shape)
print(y)
print(y.shape)

# Class distribution
unique_values = np.unique(y)
counts = np.zeros_like(unique_values)

for i, value in enumerate(unique_values):
    counts[i] = np.count_nonzero(y == value)

for i, value in enumerate(unique_values):
    print(f"Class label: {value}, Count: {counts[i]}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=40
)

print(X_train)
print(X_train.shape)
print(X_test)
print(X_test.shape)
print(y_train)
print(y_train.shape)
print(y_test)
print(y_test.shape)

# Test set class distribution
unique_values = np.unique(y_test)
counts = np.zeros_like(unique_values)

for i, value in enumerate(unique_values):
    counts[i] = np.count_nonzero(y_test == value)

for i, value in enumerate(unique_values):
    print(f"Class label: {value}, Count: {counts[i]}")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    criterion='gini',
    random_state=40
)

rf_model.fit(X_train, y_train)
preds = rf_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

classes = ["No Disease (0)", "Disease (1)"]

cm = confusion_matrix(y_test, preds)
print('Confusion Matrix')
print(cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
fig, ax = plt.subplots(figsize=(5,5))
plt.title("Confusion Matrix - Random Forest")
disp.plot(ax=ax)
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score
)

Accuracy = accuracy_score(y_test, preds)
Precision = precision_score(y_test, preds)
Recall = recall_score(y_test, preds)
F1_score = f1_score(y_test, preds)
roc_auc = roc_auc_score(y_test, preds)

print('Accuracy:', Accuracy * 100)
print('Precision:', Precision * 100)
print('Recall:', Recall * 100)
print('F1 Score:', F1_score * 100)
print('ROC AUC score: %.3f' % roc_auc)
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, preds)
auc_rf = auc(fpr, tpr)

plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
plt.plot(fpr, tpr, label='Area = {:.3f}'.format(auc_rf))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Random Forest')
plt.legend(loc='best')
plt.show()
```
